```python
from utils import  PkdbModel
import requests
import holoviews as hv
from holoviews import opts, dim
import bokeh
from bokeh.sampledata.airport_routes import routes, airports
hv.extension('bokeh')

from itertools import combinations, chain
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans
from sklearn import decomposition
import matplotlib.pyplot as plt
```

```python
studies = PkdbModel("studies", destination="1-preprocessed") 
substances = PkdbModel("substances", destination="1-preprocessed") 

substances.read()
studies.read()

substance_dict = substances.data.name.to_dict()
substance_dict_rev = {v: k for k, v in substance_dict.items()}
```

```python
substance_combinations = iter(())
for study in studies.data.itertuples():
    if isinstance(study.substances,str):
        this_substance_combinations = study.substances.split(",")
        this_substance_combinations = iter([x.strip() for x in this_substance_combinations])
        substance_combinations = chain(substance_combinations,combinations(this_substance_combinations,2))
```

```python
substances_combinations = pd.DataFrame(list(substance_combinations),columns=["substance1","substance2"])
substances_combinations["occurance"] = 1


coocurence_count = substances_combinations.groupby(["substance1","substance2"]).occurance.count().reset_index()
coocurence_count["substance1_id"] = coocurence_count["substance1"].apply(lambda x : substance_dict_rev[x])
coocurence_count["substance2_id"] = coocurence_count["substance2"].apply(lambda x : substance_dict_rev[x])
```

```python
# add substance 1 to substance 2 and vice verca
```

```python
substances_combinations2 = pd.DataFrame()
substances_combinations2["substance1"] = substances_combinations["substance2"]
substances_combinations2["substance2"] = substances_combinations["substance1"]
substances_combinations2["occurance"] = substances_combinations["occurance"]
```

```python
com = substances_combinations2.append(substances_combinations, ignore_index=True)
```

```python
coocurence_both_count = com.groupby(["substance1","substance2"]).occurance.count().reset_index()
```

```python
result = com.pivot_table(index="substance1", columns="substance2",aggfunc="sum")
res = result.fillna(0)
```

```python

transformer = Normalizer(norm="l2").fit(res) # fit does nothing.

x_trans = transformer.transform(res)
X_name = pd.DataFrame(x_trans, index=res.index, columns=res.index)
```

```python
g = sns.clustermap(X_name.T.corr(),method="single",figsize=(20,20))
```

```python
name_order = X_name.iloc[g.dendrogram_row.reordered_ind].index
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
Y = kmeans.fit_predict(X_name)
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(X_name)
X = pca.transform(X_name)
```

```python
plt.scatter(X[:,1],X[:,0], c=Y)
```

```python
X_name["color"] = Y
substances_data = substances.data.reset_index().set_index("name").loc[name_order].reset_index().rename(columns ={"substance1":"name"})
```

```python
coocurence_count["color"] = coocurence_count.substance1.apply(lambda x: X_name["color"].to_dict()[x.strip()])
```

```python
nodes = hv.Dataset(substances_data, 'index', 'name')
chord = hv.Chord((coocurence_count, nodes), ['substance1_id', 'substance2_id'], ['occurance',"color"], label="substance_combinations")
```

```python
chord.opts(
    opts.Chord(cmap='Dark2', edge_color=dim('color').str(), 
               height=800, labels='name', node_color=dim("color").str(), width=800))
```

```python
busiest = coocurence_both_count.groupby('substance1').count().sort_values('occurance', ascending=False).iloc[:30]
```

```python
busiest =list(busiest.index.values)
busiest_substances = chord.select(name=busiest, selection_mode='nodes')
```
