```python
import pandas as pd
```

```python
df = pd.read_csv("report_data.tsv", sep="\t")
df['study_count'] = [1]*len(df)

count_fields = ['study_count', 'group_count', 'individual_count', 'intervention_count', 'output_count', 'timecourse_count']
for key in count_fields:
    # cumulative sums
    df[f'{key}_cum'] = df[key].cumsum()
df
```

```python
 # 4. altair plot
import altair as alt

count_fields = ["study_count"]
for key in count_fields:
    alt.Chart(df).transform_window(
        study_count_cum='sum(study_count)',
        frame=[None, 0]
    )
    
    line = alt.Chart(df).mark_line().encode(
        x=alt.X('date:T', axis=alt.Axis(title='Date')),
        y=alt.Y(f'{key}_cum:Q', axis=alt.Axis(title='PK-DB Studies')),
        
    )
    points = alt.Chart(df).mark_circle().encode(
        alt.X('date:T'),
        alt.Y(f'{key}_cum'),
        color='group_count:N'
    )
    
points + line
```
